## Practical Problem 12

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
s, t = sp.symbols('s t')
Y = sp.Function('Y')(s)

# Define the differential equation in Laplace domain
eqn = sp.Eq(s**2 * Y - 4 * s * Y + 4 * Y, 1 +(64 * 2 / (s**2 + 2**2)))

# Solve for Y(s)
Y_s = sp.solve(eqn, Y)[0]
sp.pprint(Y_s)


```

```python
# Invert the Laplace transform to find y(t)
y_t = sp.inverse_laplace_transform(Y_s, s, t)

# Print the solution
print("The solution y(t) is:")
sp.pprint(y_t)


```

```python
# Convert the solution to a Python function
y_func = sp.lambdify(t, y_t, 'numpy')

# Generate t values for plotting
t_values = np.linspace(0, 10, 1000)

# Calculate y values using the solution function
y_values = y_func(t_values)

# Plot the graph of the solution
plt.figure(figsize=(8, 6))
plt.plot(t_values, y_values, label='y(t)')
plt.title('Solution of y" - 4y'" + 4y = 64sin(2t)")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.legend()
plt.show()
```

## Practical Problem 14

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
s, t = sp.symbols('s t')
X = sp.Function('X')(s)
Y = sp.Function('Y')(s)

# Define the differential equations in Laplace domain
eqn1 = sp.Eq(s * X + X/s, 1)  # Laplace transform of dx/dt + y = 0 with x(0) = 1
eqn2 = sp.Eq((s**2)*Y +  Y, 1)  # Laplace transform of dy/dt - x = 0 with y(0) = 0

# Solve for X(s) and Y(s)
X_s = sp.solve(eqn1, X)[0]
Y_s = sp.solve(eqn2, Y)[0]

# Invert the Laplace transforms to find x(t) and y(t)
x_t = sp.inverse_laplace_transform(X_s, s, t)
y_t = sp.inverse_laplace_transform(Y_s, s, t)

# Print the solutions
print("The solution for x(t) is:")
sp.pprint(x_t)
print("\nThe solution for y(t) is:")
sp.pprint(y_t)

```

```python
# Convert the solutions to Python functions
x_func = sp.lambdify(t, x_t, 'numpy')
y_func = sp.lambdify(t, y_t, 'numpy')

# Generate t values for plotting
t_values = np.linspace(0, 10, 1000)

# Calculate x(t) and y(t) values using the solution functions
x_values = x_func(t_values)
y_values = y_func(t_values)

# Plot the graphs of x(t) and y(t)
plt.figure(figsize=(8, 6))
plt.plot(t_values, x_values, label='x(t)')
plt.plot(t_values, y_values, label='y(t)')
plt.title('Solution of dx/dt + y = 0 and dy/dt - x = 0')
plt.xlabel('t')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


l=int(input('Periodic function of Period:'))
# Define the function f(t)
def f(t):
    if 0 < t <= l:
        return np.sin(np.pi * t / l)
    else:
        return 0

# Define the Fourier series coefficients for f(t)
def fourier_coefficient(n):
    if n  == 0:
        return 4/np.pi
    elif n % 2 == 0:
        return -4/((n**2-1)*np.pi)
    else:
        return 0

# Define the Fourier series approximation for f(t)
def fourier_series(t, N):
    series = np.zeros_like(t)
    for n in range(1, N + 1):
        series += fourier_coefficient(n) * np.cos(n * t)
    return series

# Generate t values for plotting
t_values = np.linspace(0, np.pi, 1000)

# Compute Fourier series approximation for N terms
N = 10
y_values = fourier_series(t_values, N)

# Plot the original function and its Fourier series approximation
plt.figure(figsize=(10, 6))
plt.plot(t_values, [f(t) for t in t_values], label='f(t)', color='blue')
plt.plot(t_values, y_values, label=f'Fourier series approximation (N={N})', color='red')
plt.title('Fourier Series Approximation of f(t)')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.grid(True)
plt.show()
```

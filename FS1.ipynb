## Practical Problem 1
#### Find Fourier series representing f(x)=x, 0<x<2\pi and Sketch its graph from x=-4\pi to 4\pi by python program

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the function f(x)
def f(x):
    return x

# Define the Fourier series coefficients for f(x) = x
def fourier_coefficient(n):
    if n == 0:
        return 2*np.pi
    else:
        return (-2 / n )

# Define the Fourier series approximation for f(x) = x

def fourier_series(x, N):
    series = np.zeros_like(x)
    for n in range(1, N + 1):
        series += fourier_coefficient(n) * np.sin(n * x)
    return series


# Generate x values for plotting
x_values = np.linspace(-4*np.pi, 4*np.pi, 1000)
# Compute Fourier series approximation for N terms
N = 10
y_values = fourier_series(x_values, N)

# Plot the original function and its Fourier series approximation
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='f(x) = x', color='blue')
plt.plot(x_values, y_values, label=f'Fourier series approximation (N={N})', color='red')
plt.title('Fourier Series Approximation of f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

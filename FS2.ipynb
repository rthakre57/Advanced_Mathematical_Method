```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the function f(x)
def f(x):
    return x**2

# Define the Fourier series coefficients for f(x) = x^2
def fourier_coefficient(n):
    if n == 0:
        return 2*(np.pi)**2 / 3
    else:
        return 4 *(-1)**n  / (n**2)

# Define the Fourier series approximation for f(x) = x^2
def fourier_series(x, N):
    series = np.zeros_like(x)
    for n in range(1, N + 1):
        series += fourier_coefficient(n) * np.cos(n * x)
    return series

# Generate x values for plotting
x_values = np.linspace(-np.pi, np.pi, 1000)

# Compute Fourier series approximation for N terms
N = 10
y_values = fourier_series(x_values, N)

# Plot the original function and its Fourier series approximation
plt.figure(figsize=(10, 6))
plt.plot(x_values, f(x_values), label='f(x) = x^2', color='blue')
plt.plot(x_values, y_values, label=f'Fourier series approximation (N={N})', color='red')
plt.title('Fourier Series Approximation of f(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

```
